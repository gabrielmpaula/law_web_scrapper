```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
class element_has_style(object):
  """An expectation for checking that an element has a particular css class.

  locator - used to find the element
  returns the WebElement once it has the particular css class
  """
  def __init__(self, locator, style):
    self.locator = locator
    self.style = style

  def __call__(self, driver):
    element = driver.find_element(*self.locator)   # Finding the referenced element
    if self.style in element.get_attribute("style"):
        return element
    else:
        return False

```

```python
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
driver = webdriver.Chrome(options=options)
website = 'http://www2.cjf.jus.br/jurisprudencia/trf1/'
```

```python
driver.get(website)
```

```python
pesqAvancada = driver.find_element_by_id('formulario:ckbAvancada')
pesqAvancada.click()

formText = driver.find_element_by_id('formulario:textoLivre')
formText.send_keys('colação E grau')

pesquisar = driver.find_element_by_id('formulario:actPesquisar')
pesquisar.click()

resultados = Select(driver.find_element_by_name('formulario:tabelaDocumentos_rppDD'))
resultados.select_by_value('50')
```

```python
container = soup(driver.page_source, 'lxml')
resultados = container.find_all('table', {'class':'table_resultado'})
containers = pd.read_html(str(resultados))
```

```python
next_page = driver.find_element_by_class_name('ui-paginator-next')

first_data = True

while True:
    
    container = soup(driver.page_source, 'lxml')
    resultados = container.find_all('table', {'class':'table_resultado'})
    containers = pd.read_html(str(resultados))
        
    for container in containers:
        campos = container[0::2]
        campos = list(campos[0])
        dados = container[1::2]
        dados = list(dados[0])
        df = pd.DataFrame({'campos':campos, 'dados':dados})
        df = df.set_index('campos').transpose()
        if first_data == True:
            database = df.copy()
            first_data = False
        else:
            database = database.append(df, sort=True)
   
    if next_page.get_attribute('tabindex') == '-1':
        break
    else:
        next_page.click()

    # Wait until an element with id='myNewInput' has class 'myCSSClass'
    wait = WebDriverWait(driver, 20)
    element = wait.until(element_has_style((By.ID, 'j_idt392_start'), "display: none;"))
```
